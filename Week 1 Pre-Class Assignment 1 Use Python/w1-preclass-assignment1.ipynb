<h2>[Problem 1] Create using the exponentiation arithmetic operator</h2>

```python
%%timeit
"""
Code to calculate the thickness when the paper is folded once
"""
THICKNESS = 0.00008
folded_thickness = THICKNESS*(2**43)
#print("Thickness: {} meters".format(folded_thickness))
```

<h2>[Problem 2] Unit Conversion</h2>

```python
# Convert meters to kilometers and display with two decimal places
TEN_THOUSAND_KILOMETER = 1000 * 10000
DISTANCE_TO_MOON = 384400 #in km
print("Thickness: {: .2f} ten thousand kilometers".format(folded_thickness/TEN_THOUSAND_KILOMETER))
print("Cannot reach the Moon." if folded_thickness < DISTANCE_TO_MOON*1000 else "Yes, It reached the Moon.")
```

<h2>[Problem 3] Create using a for statement</h2>

```python
%%timeit
folded_thickness = THICKNESS
for num_fold in range(0, 43):
    folded_thickness = folded_thickness * 2
#print("Thickness: {} meters".format(folded_thickness))
```

<h2>[Problem 4] Comparison of calculation time</h2>

The result is recorded when the print statement is excluded<br>
<ol>
<li>Method 1 with exponential arithmetic<br>
    <b>93.3 ns ± 1.04 ns per loop (mean ± std. dev. of 7 runs, 10000000 loops each)</b><br></li>
<li>Method 2 with for statement<br>
    <b>2.29 µs ± 44.7 ns per loop (mean ± std. dev. of 7 runs, 100000 loops each)</b></li>
</ol>

Conclusion: For statement will cost more time than the direct operation with exponential arithmetic due to the repeatation of for loop when n increases

<h2>[Problem 5] Saving to a list</h2>

```python
fold_list = list(THICKNESS*(2**i) for i in range(0, 44))
print(fold_list)
print("\nLength of list: {}".format(len(fold_list)))
```

<h2>[Problem 6] Displaying a line graph</2h>

```python
"""
Display the graph. Title and axis label name.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(fold_list) # Enter the variable name of the list in "List name"
plt.show()
```

<b>Dissusion:</b><br>
Thickness = Inital Thickness * 2^(number of folds). There is a exponential relationship between number of folds the Thickness. The rate of change is massive when it reached the end of the list.
