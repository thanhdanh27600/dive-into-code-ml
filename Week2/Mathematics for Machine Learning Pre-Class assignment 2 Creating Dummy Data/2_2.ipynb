<h2>[Problem 1] Creating random numbers</h2>

```python

import numpy as np

mean = [-3, 0]
covariance = [[1.0 , 0.8 ],
                [0.8, 1.0]]
# To ensure that the value does not change from one run to the next
np.random.seed(0)
random_numbers  = np.random.multivariate_normal(mean, covariance, 500)
```

```python
# uncomment to show numbers
#print(random_numbers)
```

<h2> [Problem 2] Visualization by scatter plot <h2>

```python
import matplotlib.pyplot as plt
# ":" is choose all row
# ":,0" with all rows, choose column 0 for the x axis
# ":,1" with all rows, choose column 1 for the y axis 
plt.scatter(random_numbers[:,0],random_numbers[:,1])
```

<h2>[Problem 3] Visualization by histogram <h2>

```python
# 50 is number of "bars", larger means the range of each value is narrower
# range of [-6:3] fits 2 histogram perfectly!
plt.hist(random_numbers[:,0], 50)
plt.xlim(left=-6, right=3)
plt.show()
plt.hist(random_numbers[:,1], 50)
plt.xlim(left=-6, right=3)
plt.show()
```

<h2>[Problem 4] Addition of data <h2>

```python
mean2 = [0, -3]
covariance2 = [[1.0 , 0.8 ],
                [0.8, 1.0]]
# To ensure that the value does not change from one run to the next
np.random.seed(0)
random_numbers2  = np.random.multivariate_normal(mean2, covariance2, 500)
ax = plt.figure().add_subplot(111)
ax.scatter(random_numbers[:,0],random_numbers[:,1],label='0')
ax.scatter(random_numbers2[:,0],random_numbers2[:,1],label='1')
plt.legend()
plt.show()
```

<h2>[Problem 5] Data combination</h2>

```python
array_concatenated = np.concatenate((random_numbers,random_numbers2))
```

```python
# Well, it is the same color when use append. We should label it first.
plt.scatter(array_concatenated[:,0], array_concatenated[:,1])
plt.show()
```

<h2>[Problem 6] Labeling</h2>

```python
labeled_array1 = np.insert(random_numbers, 2, 0, axis=1)
labeled_array2 = np.insert(random_numbers2, 2, 1, axis=1)
array_concatenated = np.concatenate((labeled_array1,labeled_array2))
plt.scatter(array_concatenated[:,0], array_concatenated[:,1], c=array_concatenated[:,2])
```
