<h2>[Problem 1] Number of grains of rice on the 100th day<h2>

<h3>Firstly, I found the easiest way and also the most effective way to calculate the total grains from beginning to that day, which is:<br>
<blockquote>At a certain day d (d â‰¥ 0), there will be <b>2^d</b> grains, and total grains equals <b>1+2+2^2+2^3+...+2^d = 2^(d+1)-1<b><br></blockquote>
Therefore it is easier to use exponential operation rather than for statement to calculate the whole sum.<h3>

```python
# Calculate and print list_n_grains and list_total_grains
# there will be 2^(day-1) grain of rice.
# The assignment with 2 list at the same time works as tuple, this is for shortening.
list_n_grains, list_total_grains = list(2**i for i in range(0,100)), list(2**(i+1) - 1 for i in range(0,100))
# Print out how many grains of rice you can get for the number of days you take the course.
print("Grains of rice in 100th day: {}".format(list_n_grains[-1]))
print("The total number of rice grains from day 1 to day 100: {}".format(list_total_grains[-1]))
# Plot list_n_grains
plt.subplot(311)
plt.title("Number of rice in day 100")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.plot(list_n_grains, color='#0062ff')

# Plot list_total_grains
plt.subplot(313)
plt.title("Total rice over 100 days")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.plot(list_total_grains, color='#fc0303')
```

<h2>[Problem 2] Number of rice grains outside of the 100th day<h2>

```python
def compute_sorori_shinzaemon(day=100):
   """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameters
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    list_n_grains, list_total_grains = list(2**i for i in range(0,day)), list(2**(i+1) - 1 for i in range(0,day))
    return list_n_grains, list_total_grains
```

```python
# Course duration is 5 Weeks Pre-class and 3 Months Full-time program, made it approximately total 5*7 + 3*30 = 118 days.
COURSE_DURATION = 118
# Call the funtion with day = duration of the course
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=COURSE_DURATION)
# Print out how many grains of rice you can get for the number of days you take the course.
print("Grains of rice in last day of this course: {}".format(list_n_grains[-1]))
print("The total number of rice grains during this course: {}".format(list_total_grains[-1]))
# Plot list_n_grains
plt.subplot(311)
plt.title("Number of rice in last day of this course")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.plot(list_n_grains, color='#0062ff')

# Plot list_total_grains
plt.subplot(313)
plt.title("Total rice during this course")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.plot(list_total_grains, color='#fc0303')
```

<h2>[Problem 3] How many people can live for how many days with the rice we get?<h2>

<h3>How many grants of rice a Vietnamese consumes per day  (in average)?</h3>
I have done some researches. According to <b>Reference 1</b>. A Vietnamese person consumes <b>96.6</b> kg of rice / year.<br>
There are roughly 365 days in a year. Then 1 Vietnamese consumes <b>(96.6)/365 * 1000 ~ 264.66 gram grain per day.</b>
<img src="./2.JPG"><br>
Then convert the gram of grains to number of grains rice, I use <b>Reference 2</b>. The result is GRANTS_PER_PERSON_PER_DAY = 4084 <br>
<img src="./3.JPG">


```python
# The result of how many grants of rice a Vietnamese consume per day (as above)
GRANTS_PER_PERSON_PER_DAY = 4084

def days_to_live(number_of_rice_grains, number_of_people):
     """
    A function that calculate how many people can live for how many days with this rice. 
    Parameters
    --------------
    number_of_rice_grains: int
        Number of rice grains to be alive (consumed)
    number_of_people: int
        Number of people consumes the rice grains
    Returns
    --------------
    number_of_days_to_live : long long int
        Number of days that people can survive by consumed the number of rice grains
    
    """
    # Each day, the total grants consumed is number_of_people * GRANTS_PER_PERSON_PER_DAY. Hence, let the total rice grains divided by grants consumed per day.
    number_of_days_to_live = number_of_rice_grains/(number_of_people * GRANTS_PER_PERSON_PER_DAY)
    return number_of_days_to_live
```

```python
# This course contains 41 students but 11 of who are 4-year students has been removed.
NUMBER_OF_STUDENTS = 30
# Pass the total grains the previous question and number of student, then print the result
print(days_to_live(list_total_grains[-1], NUMBER_OF_STUDENTS), " days")
```
