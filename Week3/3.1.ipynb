# [Problem 6] Creating a diagram

```python
species = df['Species'].value_counts()
species_index = iris.target_names
species_value = species.values
#explore = (0,0.2,0)
plt.pie(species_value,colors = ['c', 'orange', 'violet'], autopct = '%1.1f%%', explode=[0,0,0.1],startangle=90)
plt.xlabel('species')
plt.legend(labels=species_index, loc='lower right')
plt.show()
```

```python
labels = iris.feature_names
for i in range(len(labels)):
    df.boxplot([labels[i]])
    plt.show()
```

In general, violin plots can be considered a combination of the box plot with a kernel density plot. In the violin plot, we can find the same information as in the box plots:
*  Median (a white dot on the violin plot)
*  Interquartile range (the black bar in the center of violin)
*  The lower/upper adjacent values (the black lines stretched from the bar)<br>

<img src="1.PNG"/><br>

Advantage of the violin plot over the box plot is that aside from showing the abovementioned statistics it also shows the entire distribution of the data. This is of interest, especially when dealing with multimodal data, i.e., a distribution with more than one peak.

```python
import seaborn
sea_iris = seaborn.load_dataset('iris')
```

```python
seaborn.pairplot(sea_iris, hue='species')
plt.show()
```

```python
# Correlation matrix
iris_correlation = sea_iris.corr()
iris_correlation
```

```python
seaborn.heatmap(iris_correlation, cmap=plt.cm.RdYlGn)
plt.show()
```

**Looking at the heat map of the correlation matrix, it can be seen that the petal length and petal width have a high correlation of 0.96. From the correlation diagram, the sepal length, petal length, and thickness can also have a correlation of 0.8 or more. From the readable boxplot and violin plot, it can be read that the sepal length, petal length, and petal width increase in the order of setosa<virgicolr<virgnica, but setosa gives a higher numerical value for the sepal width. From this, it can be inferred that setosa is a flower with short sepals and wide width.**
